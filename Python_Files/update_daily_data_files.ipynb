```python
import pandas as pd
import os
from datetime import datetime
```

```python
# Define the folder containing the files
folder_path = "../Folder 3/"

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
```

```python
# Load data from other CSV files (unchanged)
citizens_round2 = pd.read_csv("../Data_Source/Retro Round 2.csv", encoding='latin1')
citizens_round3 = pd.read_csv("../Data_Source/Retro Round 3.csv", encoding='latin1')
citizens_round4 = pd.read_csv("../Data_Source/Retro Round 4.csv", encoding='latin1')
grants = pd.read_csv("../Data_Source/Grants_Council.csv", encoding='latin1')
grants_mm_season5 = pd.read_csv("../Data_Source/Grants_Council_MM.csv", encoding='latin1')
dab_season5 = pd.read_csv("../Data_Source/Developer_Advisory_Board.csv", encoding='latin1')
coc_season5 = pd.read_csv("../Data_Source/Code_of_Conduct_Council.csv", encoding='latin1')
```

```python
# Drop rows with null values in the 'address' column for each DataFrame
citizens_round2.dropna(subset=['address'], inplace=True)
citizens_round3.dropna(subset=['address'], inplace=True)
citizens_round4.dropna(subset=['address'], inplace=True)
grants.dropna(subset=['address'], inplace=True)
grants_mm_season5.dropna(subset=['address'], inplace=True)
dab_season5.dropna(subset=['address'], inplace=True)
coc_season5.dropna(subset=['address'], inplace=True)
```

```python
# Filter grants data by season (unchanged)
grants_season3 = grants[grants['season'] == 3]
grants_season4 = grants[grants['season'] == 4]
grants_season5 = grants[grants['season'] == 5]
```

```python
# Convert start_date and end_date columns to datetime for comparison
citizens_round2['start_date'] = pd.to_datetime(citizens_round2['start_date'], format='%m/%d/%Y')
citizens_round2['end_date'] = pd.to_datetime(citizens_round2['end_date'], format='%m/%d/%Y')
citizens_round3['start_date'] = pd.to_datetime(citizens_round3['start_date'], format='%m/%d/%Y')
citizens_round3['end_date'] = pd.to_datetime(citizens_round3['end_date'], format='%m/%d/%Y')
citizens_round4['start_date'] = pd.to_datetime(citizens_round4['start_date'], format='%m/%d/%Y')
citizens_round4['end_date'] = pd.to_datetime(citizens_round4['end_date'], format='%m/%d/%Y')
grants_season3['start_date'] = pd.to_datetime(grants_season3['start_date'], format='%m/%d/%Y')
grants_season3['end_date'] = pd.to_datetime(grants_season3['end_date'], format='%m/%d/%Y')
grants_season4['start_date'] = pd.to_datetime(grants_season4['start_date'], format='%m/%d/%Y')
grants_season4['end_date'] = pd.to_datetime(grants_season4['end_date'], format='%m/%d/%Y')
grants_season5['start_date'] = pd.to_datetime(grants_season5['start_date'], format='%m/%d/%Y')
grants_season5['end_date'] = pd.to_datetime(grants_season5['end_date'], format='%m/%d/%Y')
grants_mm_season5['start_date'] = pd.to_datetime(grants_mm_season5['start_date'], format='%m/%d/%Y')
grants_mm_season5['end_date'] = pd.to_datetime(grants_mm_season5['end_date'], format='%m/%d/%Y')
dab_season5['start_date'] = pd.to_datetime(dab_season5['start_date'], format='%m/%d/%Y')
dab_season5['end_date'] = pd.to_datetime(dab_season5['end_date'], format='%m/%d/%Y')
coc_season5['start_date'] = pd.to_datetime(coc_season5['start_date'], format='%m/%d/%Y')
coc_season5['end_date'] = pd.to_datetime(coc_season5['end_date'], format='%m/%d/%Y')
```

```python
# Helper function to add missing delegates (unchanged)
def add_missing_delegates(data, new_addresses):
    missing_addresses = set(new_addresses['address'].str.lower()) - set(data['delegate'].str.lower())
    missing_df = pd.DataFrame({
        'delegate': list(missing_addresses),
        'voting_power': 0
    })
    data = pd.concat([data, missing_df], ignore_index=True)
    return data
```

```python
# Process each file in the folder
for file_name in csv_files:
    file_path = os.path.join(folder_path, file_name)
    
    # Load the daily delegate data
    data = pd.read_csv(file_path)

    # Extract the date from the file name (assumed format 'YYYY-MM-DD.csv')
    file_date = datetime.strptime(file_name[:-4], '%Y-%m-%d')

    # Add missing delegates from various rounds and councils (unchanged)
    data = add_missing_delegates(data, citizens_round2)
    data = add_missing_delegates(data, citizens_round3)
    data = add_missing_delegates(data, citizens_round4)
    data = add_missing_delegates(data, grants_season3)
    data = add_missing_delegates(data, grants_season4)
    data = add_missing_delegates(data, grants_season5)
    data = add_missing_delegates(data, grants_mm_season5)
    data = add_missing_delegates(data, dab_season5)
    data = add_missing_delegates(data, coc_season5)
    
    # Add the columns (unchanged)
    data['th_vp'] = None
    data['ch_member_r2'] = None
    data['ch_vp_r2'] = None
    data['ch_member_r3'] = None
    data['ch_vp_r3'] = None
    data['ch_member_r4'] = None
    data['ch_vp_r4'] = None
    data['gc_member_s3'] = None
    data['gc_vp_s3'] = None
    data['gc_member_s4'] = None
    data['gc_vp_s4'] = None
    data['gc_member_s5'] = None
    data['gc_vp_s5'] = None
    data['gc_member_mm_s5'] = None
    data['gc_vp_mm_s5'] = None
    data['sc_member_s5'] = None
    data['sc_vp_s5'] = None
    data['coc_member_s5'] = None
    data['coc_vp_s5'] = None
    data['dab_member_s5'] = None
    data['dab_vp_s5'] = None

    # Create membership columns for existing delegates only if the round/council is active on the file_date
    data['ch_member_r2'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round2['start_date'].iloc[0] and 
                  file_date <= citizens_round2['end_date'].iloc[0] and
                  x.lower() in citizens_round2['address'].str.lower().values else 0)

    data['ch_member_r3'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round3['start_date'].iloc[0] and 
                  file_date <= citizens_round3['end_date'].iloc[0] and
                  x.lower() in citizens_round3['address'].str.lower().values else 0)

    data['ch_member_r4'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round4['start_date'].iloc[0] and 
                  file_date <= citizens_round4['end_date'].iloc[0] and
                  x.lower() in citizens_round4['address'].str.lower().values else 0)

    data['gc_member_s3'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_season3['start_date'].iloc[0] and 
                  file_date <= grants_season3['end_date'].iloc[0] and
                  x.lower() in grants_season3['address'].str.lower().values else 0)

    data['gc_member_s4'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_season4['start_date'].iloc[0] and 
                  file_date <= grants_season4['end_date'].iloc[0] and
                  x.lower() in grants_season4['address'].str.lower().values else 0)

    data['gc_member_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_season5['start_date'].iloc[0] and 
                  file_date <= grants_season5['end_date'].iloc[0] and
                  x.lower() in grants_season5['address'].str.lower().values else 0)

    data['gc_member_mm_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_mm_season5['start_date'].iloc[0] and 
                  file_date <= grants_mm_season5['end_date'].iloc[0] and
                  x.lower() in grants_mm_season5['address'].str.lower().values else 0)

    data['dab_member_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= dab_season5['start_date'].iloc[0] and 
                  file_date <= dab_season5['end_date'].iloc[0] and
                  x.lower() in dab_season5['address'].str.lower().values else 0)

    data['coc_member_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= coc_season5['start_date'].iloc[0] and 
                  file_date <= coc_season5['end_date'].iloc[0] and
                  x.lower() in coc_season5['address'].str.lower().values else 0)

    # Calculate and assign voting power percentages based on active membership
    sum_th_vp = data[data['voting_power'] > 1]['voting_power'].sum()
    data['th_vp'] = data.apply(lambda row: (row['voting_power'] * 100) / sum_th_vp if row['voting_power'] > 1 else 0, axis=1)

    count_ch_member_r2 = data['ch_member_r2'].sum()
    data['ch_vp_r2'] = data.apply(lambda row: (row['ch_member_r2'] * 100) / count_ch_member_r2 if row['ch_member_r2'] == 1 else 0, axis=1)

    count_ch_member_r3 = data['ch_member_r3'].sum()
    data['ch_vp_r3'] = data.apply(lambda row: (row['ch_member_r3'] * 100) / count_ch_member_r3 if row['ch_member_r3'] == 1 else 0, axis=1)

    count_ch_member_r4 = data['ch_member_r4'].sum()
    data['ch_vp_r4'] = data.apply(lambda row: (row['ch_member_r4'] * 100) / count_ch_member_r4 if row['ch_member_r4'] == 1 else 0, axis=1)

    count_gc_member_s3 = data['gc_member_s3'].sum()
    data['gc_vp_s3'] = data.apply(lambda row: (row['gc_member_s3'] * 100) / count_gc_member_s3 if row['gc_member_s3'] == 1 else 0, axis=1)

    count_gc_member_s4 = data['gc_member_s4'].sum()
    data['gc_vp_s4'] = data.apply(lambda row: (row['gc_member_s4'] * 100) / count_gc_member_s4 if row['gc_member_s4'] == 1 else 0, axis=1)

    count_gc_member_s5 = data['gc_member_s5'].sum()
    data['gc_vp_s5'] = data.apply(lambda row: (row['gc_member_s5'] * 100) / count_gc_member_s5 if row['gc_member_s5'] == 1 else 0, axis=1)

    count_gc_member_mm_s5 = data['gc_member_mm_s5'].sum()
    data['gc_vp_mm_s5'] = data.apply(lambda row: (row['gc_member_mm_s5'] * 100) / count_gc_member_mm_s5 if row['gc_member_mm_s5'] == 1 else 0, axis=1)

    count_dab_member_s5 = data['dab_member_s5'].sum()
    data['dab_vp_s5'] = data.apply(lambda row: (row['dab_member_s5'] * 100) / count_dab_member_s5 if row['dab_member_s5'] == 1 else 0, axis=1)

    count_coc_member_s5 = data['coc_member_s5'].sum()
    data['coc_vp_s5'] = data.apply(lambda row: (row['coc_member_s5'] * 100) / count_coc_member_s5 if row['coc_member_s5'] == 1 else 0, axis=1)

    # Save the updated data back to the same file (overwrite)
    data.to_csv(file_path, index=False)

    # Print confirmation message for each file
    print(f"File {file_name} updated successfully!")

print("All files updated successfully!")
```
