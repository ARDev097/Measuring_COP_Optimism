```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("../Data_Source/data.csv", encoding='latin1')
```

## CPI by squaring the total_influence_delegates

```python
# Calculate the squared total voting power
data['influence_r2_s3_squared'] = data['influence_r2_s3'] ** 2

# Calculate the mHHI
mHHI = round(data['influence_r2_s3_squared'].sum(), 6)
print("mHHI for Round 2 and Season 3:", mHHI)
```

```python
# Calculate the squared total voting power
data['influence_r2_s4_squared'] = data['influence_r2_s4'] ** 2

# Calculate the mHHI
mHHI = round(data['influence_r2_s4_squared'].sum(), 6)
print("mHHI for Round 2 and Season 4:", mHHI)
```

```python
# Calculate the squared total voting power
data['influence_r3_s5_squared'] = data['influence_r3_s5'] ** 2

# Calculate the mHHI
mHHI = round(data['influence_r3_s5_squared'].sum(), 6)
print("mHHI for Round 3 and Season 5:", mHHI)
```

```python
# Calculate the squared total voting power
data['influence_r4_s5_squared'] = data['influence_r4_s5'] ** 2

# Calculate the mHHI
mHHI = round(data['influence_r4_s5_squared'].sum(), 6)
print("mHHI for Round 4 and Season 5:", mHHI)
```

```python
# Data
labels = ['Token_House', 'Round2_Season3', 'Round2_Season4', 'Round2_Season5', 'Round3_Season5']
mhhi_values = [318.78, 137.28, 137.28, 90.03, 79.59]
colors = ['purple', 'red', 'blue', 'green', 'orange']

# Bubble sizes (proportional to HHI values)
bubble_sizes = [value * 10 for value in mhhi_values]  # Scaling up for better visibility

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(labels, mhhi_values, s=bubble_sizes, alpha=0.4, c=colors, edgecolors="w", linewidth=2)

# Annotate each bubble with the HHI value
for i in range(len(labels)):
    plt.text(labels[i], mhhi_values[i], f'{mhhi_values[i]:.2f}', ha='center', va='center', fontsize=9, color='black', weight='bold')

# Labels and title
plt.xlabel('CPI in different Retro Rounds and Seasons')
plt.ylabel('CPI Values')
plt.title('CPI Values of Optimism')

# Set y-axis range
plt.ylim(0, 500)

# Display the chart
plt.show()
```
